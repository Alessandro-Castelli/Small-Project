```python
# Create a mini-dataset for each column.
mini_datasets = {}
for column in df4.columns:
    mini_datasets[column] = pd.DataFrame(df4[column])

total_entries = 262968

# Create a time index for 30 years, with a time unit of one hour each.
time_index = pd.date_range(start="01-01-1986", periods=total_entries, freq='H')

# The number of years covered by the dataset.
total_years = 30

# The number of rows per year.
entries_per_year = total_entries // total_years

# Create and display time series charts for each column with a year-wise breakdown
for column, mini_dataset in mini_datasets.items():
    plt.figure(figsize=(50, 30))

    # Calculate the number of rows for each chart (year)
    rows = math.ceil(total_years / 2)
    columns = 2

    for year in range(total_years):
        plt.subplot(rows, columns, year + 1)
        start_idx = year * entries_per_year
        end_idx = (year + 1) * entries_per_year

        plt.plot(time_index[start_idx:end_idx], mini_dataset[start_idx:end_idx])
        plt.title(f"YEAR {year + 1}")
        plt.xlabel("TIME")
        plt.ylabel(column)
        plt.grid()

    plt.suptitle(f"TEMPORAL GRAPHICS: {column}")
    plt.tight_layout(rect=[0, 0, 1, 0.95])

    #plt.show()
    output_directory = "/content/Table1"
    output_file = os.path.join(output_directory, f"{column}_temporal_graph.png")
    plt.savefig(output_file)

plt.close('all')
```

```python
# Initialize a list to store the average values for each year
average_values = {column: [] for column in mini_datasets}

# Calculate the average values for each year
for year in range(total_years):
    start_idx = year * entries_per_year
    end_idx = (year + 1) * entries_per_year
    for column, mini_dataset in mini_datasets.items():
        yearly_data = mini_dataset[start_idx:end_idx]
        average_value = yearly_data.mean().values[0]
        average_values[column].append(average_value)

# Create a graph of the average values for each year
plt.figure(figsize=(12, 6))
for column, values in average_values.items():
    plt.plot(range(1986, 2016), values, label=column)

plt.title("Average Solar Energy Production per Year")
plt.xlabel("Year")
plt.ylabel("Average Solar Energy Production")
plt.legend(loc="upper right")
plt.grid()

plt.show()
```

#Decennial calculation of mean values for each country

```python
# Calculate the length of each third of the column
third_len = len(df4) // 3

# Create an empty dictionary to store the results
means = {}

# For each column in the dataset
for col in df4.columns:
    # I want to calculate the mean for each third of the column
    means[col] = [
        df4[col][:third_len].mean(),
        df4[col][third_len:2*third_len].mean(),
        df4[col][2*third_len:].mean()
    ]

# Print the avereges
for col, means_third in means.items():
    print(f"Means for the column {col}: {means_third}")

# Create an array for the x-axis
decadi = ['First decade', 'Second decade', 'Third decade']

# Create a new picture
plt.figure(figsize=(18, 10))

# Create a marker list
markers = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']

# For each column in the dictionary of averages
for i, (col, means_third) in enumerate(means.items()):
    # Create a line chart with the averages of the thirdS
    plt.plot(decadi, means_third, marker=markers[i % len(markers)], label=col)

# Set the chart title
plt.title('Average per decade')

# Set the axis labels
plt.xlabel('Decade')
plt.ylabel('Mean')

# Set the caption
plt.legend()

# Show
plt.show()
```
